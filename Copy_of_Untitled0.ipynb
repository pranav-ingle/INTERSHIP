```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (use exact file name)
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)

# View first few rows to check data
print(df.head())

# View column names
print(df.columns)
```

```python
# Import libraries (if not already imported)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# If your dataset is already loaded as 'df', skip loading again

# Replace NaN with 0 for plotting purposes (optional)
df['2024'] = df['2024'].fillna(0)

# Plot Histogram for 2024 population
plt.figure(figsize=(10,6))
sns.histplot(df['2024'], bins=20, kde=True, color='skyblue')
plt.title('Population Distribution in 2024')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()

# Plot Bar chart for top 10 countries by population in 2024
top10 = df[['Country Name', '2024']].sort_values(by='2024', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='2024', y='Country Name', data=top10, palette='viridis')
plt.title('Top 10 Countries by Population in 2024')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
